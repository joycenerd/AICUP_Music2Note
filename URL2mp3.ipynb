```python
import os
file_names = os.listdir('./Data/MIR')
file_names.sort()
file_names.remove('AIcup提供dataset說明.docx')
file_names.remove('traverse.py')
try:
    file_names.remove('MIR')
except:
    None
print (file_names)
```

```python
from pytube import YouTube
import shutil
import threading
from time import sleep

sem=threading.Semaphore(4)
def download(url, name, total):
    with  sem: 
        try:
            yt = YouTube(url)
            yt = yt.streams.filter(only_audio=True).first()
            yt.download(output_path='./Data/MIR/{}'.format(name), filename=name)
            shutil.move('./Data/MIR/'+ name + '/' + name +'.mp4','./Data/MIR/' + name + '/' + name + '.mp3')
            global now
            print("[{}/{}] {}".format(now, total,name))
            now += 1 
        except:
            None
now = 1
while (len(file_names) != 0) :
    threads = []
    for name in file_names:
        with open('./Data/MIR/{}/{}_link.txt'.format(name,name)) as file:
            url = file.read()
        thread = threading.Thread(target = download,args = (url,name,len(file_names)))  
        thread.start() 
        threads.append(thread)
        sleep(0.2)

    for thread in threads:
        thread.join() 

    for name in file_names:
        try:
            open('./Data/MIR/{}/{}.mp3'.format(name,name))
            file_names.remove(name)
        except:
            with open('./Data/MIR/{}/{}_link.txt'.format(name,name)) as file:
                url = file.read()
            print ('Download again ',name ,url)

            
```
